```python
import numpy as np
# matplotlib
import matplotlib.pyplot as plt
```

```python
year = [2019,2020,2021,2022]
gdp_india = [2.83,2.67,3.15,3.38]
# line plot - used to show values over time
plt.plot(year,gdp_india)
plt.show()
plt.clf() # clears the plot
plt.plot(year,gdp_india)
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{int(x)}"))
plt.xlabel('Year')
plt.ylabel('India\'s GDP')
plt.title("India's GDP for last 4 years")
plt.yticks([2.83,2.67,3.15,3.38],['2.83T','2.67T','3.15T','3.38T'])
plt.show()
```

```python
# scatter plot - used to show correlation between x and y variables
pop = [1383,1396,1407,1417]
plt.scatter(pop,gdp_india)
plt.show()
plt.clf()

# other parameters, s: size , c: color , alpha: opacity of dots
# plt.text(x,y,'value')
# plt.grid(True) : to show the value at intersection of x & y data points in the grid

plt.scatter(pop,gdp_india, s = np.array(pop)/4, c = 'Red', alpha = 0.5)
plt.xlabel('Population in millions')
plt.ylabel('India\'s GDP in trillions')
plt.title("India's GDP for last 4 years with population growth")
plt.show()
```

```python
# histogram - used to show distribnution of data
# Too few bins will oversimplify reality and won't show you the details. 
# Too many bins will overcomplicate reality and won't show the bigger picture

marks = [45,60,99,87,72,91,54,89,90,34]
plt.hist(marks, bins =10)
plt.xlabel('Marks')
plt.ylabel('Count of Students')
plt.title("Score Analysis")
plt.show()
```

```python
# Dictionaries: they store data in a key: value pair
```

```python
family = {'Baba':73, 'Bhai':35, 'Me':33}
print(family)
print(family.keys())
family['maa'] = 42
print(family)
del(family['maa'])
print(family)
```

```python
#keys are unique. If multiple values for the same key, last value will be stored.
family = {'Baba':73, 'Bhai':35, 'Me':33, 'Baba':75}
print(family)
```

```python
# keys are immutable objects. So, lists can't be keys. Lists can be values though. Even dictonaries can be values also.
family = {'Baba':{'age':75,'job':'bank'},
          'Bhai':{'age':33,'job':'it'}}
print(family)
print(family['Baba']['job'])
```

```python
family = {'Baba':[75,'bank'],
          'Bhai':[33,'it']}
print(family)
print(family['Baba'][0])
```

```python
DF = pd.read_csv('liver_cancer_prediction.csv')
print(DF)
```

```python
# index & select data
```

```python
# 1. Square brackets
```

```python
print(DF['Country']) 
print(type(DF['Country'])) # single bracket, series
```

```python
print(type(DF[['Country']])) # double bracket, DataFrame
```

```python
print(DF[['Country','Region']])
print(type(DF[['Country','Region']]))
```

```python
print(DF[1:4]) # slciing using interger index
# print(DF[1:3][2])
# print(DF[1:3,2])
# Throws error, so, we use loc & iloc, for subsetting like numpy 2-D array.
```

```python
# loc
```

```python
print(DF.loc[1]) # one row
print(type(DF.loc[1])) # single bracket, series
```

```python
print(DF.loc[[1]]) # one row
print(type(DF.loc[[1]])) # double bracket, DataFrame
```

```python
print(DF.loc[[1,2],['Country','Region']]) # 2 rows 2 columns
```

```python
print(DF.loc[:,['Country','Region']]) # all rows + 2 columns
```

```python
# iloc : instead of column names, index used
```

```python
print(DF.iloc[1]) # one row as series
```

```python
print(DF.iloc[[1]]) # one row, double bracket, DataFrame
```

```python
print(DF.iloc[[1,2]]) # 2 rows as data frame
```

```python
print(DF.iloc[[1,2],[0,1]]) # 2 rows & 2 columns
```

```python
print(DF.iloc[:,[0,1]]) # all rows and 2 columns
```
