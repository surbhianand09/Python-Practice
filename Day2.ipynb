```python
import numpy as np
# matplotlib
import matplotlib.pyplot as plt
```

```python
year = [2019,2020,2021,2022]
gdp_india = [2.83,2.67,3.15,3.38]
# line plot - used to show values over time
plt.plot(year,gdp_india)
plt.show()
plt.clf() # clears the plot
plt.plot(year,gdp_india)
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{int(x)}"))
plt.xlabel('Year')
plt.ylabel('India\'s GDP')
plt.title("India's GDP for last 4 years")
plt.yticks([2.83,2.67,3.15,3.38],['2.83T','2.67T','3.15T','3.38T'])
plt.show()
```

```python
# scatter plot - used to show correlation between x and y variables
plt.scatter(year,gdp_india)
plt.show()
plt.clf()

# other parameters, s: size , c: color , alpha: opacity of dots
# plt.text(x,y,'value')
# plt.grid(True) : to show the value at intersection of x & y data points in the grid
pop = [1383,1396,1407,1417]
plt.scatter(year,gdp_india, s = np.array(pop)/4, c = 'Red', alpha = 0.5)
plt.xlabel('Year')
plt.ylabel('India\'s GDP')
plt.title("India's GDP for last 4 years with population growth")
plt.show()
```

```python
# histogram - used to show distribnution of data
# Too few bins will oversimplify reality and won't show you the details. 
# Too many bins will overcomplicate reality and won't show the bigger picture

marks = [45,60,99,87,72,91,54,89,90,34]
plt.hist(marks, bins =10)
plt.xlabel('Marks')
plt.ylabel('Count of Students')
plt.title("Score Analysis")
plt.show()
```
